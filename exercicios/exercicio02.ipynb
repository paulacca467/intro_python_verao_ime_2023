```python
import pandas as pd
```

```python
df = pd.read_csv('../dados/titanic.csv')
df
```

```python
#Q1
df['sobreviveu'].value_counts()
```

```python
#Q2
df[['nome']].loc[(df['idade'] < 12) & (df['sobreviveu'] == 'sim')].count()


```

```python
#Q3
embarque = df.loc[~df['embarque'].isin(['Cherbourg'])]
embarque[['embarque']].value_counts()
```

```python
#Q4
passageiras = df.loc[df['sexo'].isin(['feminino'])]
passageiras
```

```python
#Q5
criancas = df.loc[(df['idade'] < 12)]
criancas_cherbourg = criancas.loc[criancas['embarque'].isin(['Cherbourg'])]
criancas_cherbourg
```

```python
#Q6
titanic_menor = df[['nome', 'classe', 'idade', 'embarque']].copy()
titanic_menor
```
